```python
import numpy as np 
import pandas as pd
import ast
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#Removing Unescessary Columns
#Columns to be Removed:Budget,Homepage,original_title,original_language,popularity,production_companies,production_countries,release_date,revenue,runtime,spoken_languagues,status,tagline,vote_average,vote_count,movie_id
#Columns We are Keeping:genre,id,keywords,title,Overview,Cast,Crew
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#Merging Overview,Genres,Keywords,Cast,Crew into Tags
#First Preprocessing for the Columns
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
#We need to convert this to a format with all the genres in a single list
#list is in form of string tho so we use literal eval to convert it to normal list

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#same process for keywords
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#Same for cast but only top 3 cast members
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
#function to get only director names
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#converting overview from string to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Need to remove spaces  in the lists  as ppl might have same first names 
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#Make new column tag and put the 4 lists concated into it
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#Converting tags into String
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#converting tags to lowercase
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
#Text Vectorziation
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Max words is 5000 as we  take the 5000 most common words after combining all the tags
#Check the frequency of these words in the tags of each movie
#Removing Stop words like in on is etc that have no meaning  to the movie
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#making each movie as a vector
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#aLL Similar words stored under the same tag for ex love loved loving etc..

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#We will use cosine distance to compare distance between movies as it is a high dimensional comparison
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#Now making the recomendation function
def recommend(movie):
    movie_index= new_df[new_df['title']== movie].index[0]
    #Getting the distances between each movie and the given movie
    distances=similarity[movie_index]
    #We use enumurate so we dont lose the index position when sorting to get most similar movies
    #We use reverse of sorting to get descendeding order,we use lwe use key to sort based on the similairt and not the index number
    #We use [1:6] so we dont consider the movie itslef but get its 5 most similar movies
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
